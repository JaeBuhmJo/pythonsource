```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

import warnings
warnings.simplefilter("ignore")

sns.set_theme()
sns.set(rc={"figure.figsize":(10,5)})

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
# cctv 데이터 로드
cctv_df = pd.read_csv("./data/CCTV 설치 현황_221231기준.csv",encoding="cp949", thousands=",",skiprows=[1])
cctv_df.head(3)
```

```python
cctv_df = cctv_df[['구분','총계','2013년 이전\n설치된 CCTV','2013년','2014년','2015년','2016년','2017년','2018년','2019년','2020년','2021년','2022년']]
```

```python
cctv_df.head(1)
```

```python
cctv_df.rename(columns={"구분":"구 별","총계":"cctv 총계","2013년 이전\n설치된 CCTV":"2013년 이전"},inplace=True)
```

```python
cctv_df
```

```python
cctv_df["2020년 이전"] = cctv_df["2013년 이전"]+cctv_df["2014년"]+cctv_df["2015년"]+cctv_df["2016년"]+cctv_df["2017년"]+cctv_df["2018년"]+cctv_df["2019년"]+cctv_df["2020년"]+cctv_df["2021년"]+cctv_df["2022년"]
```

```python
cctv_df
```

```python
cctv_df.drop(columns=['2020년 이전'],inplace=True)
```

```python
cctv_df["2020년 이전"]=cctv_df[cctv_df.columns.difference(["구 별","cctv 총계","2021년","2022년"])].sum(axis=1)
cctv_df
```

```python
cctv_df.drop(columns=['2013년 이전', '2013년', '2014년', '2015년', '2016년', '2017년', '2018년', '2019년', '2020년'],inplace=True)
```

```python
cctv_df.head(3)
```

```python
# 중 구 ==> 중구
cctv_df.iloc[1,0] = "중구"
cctv_df
```

```python
# 데이터 분석
# 1. 구별 cctv가 가장 적게 설치되어있는 구
print(cctv_df.sort_values("cctv 총계").head(1))
print(cctv_df.sort_values("cctv 총계").tail(1))
```

```python
seoul_df = pd.read_excel("./data/주민등록인구(월별)_202212.xlsx",skiprows=[0,1,3], usecols="B,D,G,J")
seoul_df
```

```python
seoul_df.columns=["구 별","총인구","한국인","외국인"]
seoul_df
```

```python
seoul_df.set_index("구 별")
```

```python
seoul_df.sort_values("외국인",ascending=False)
```

```python
merge_df = pd.merge(seoul_df,cctv_df,on="구 별")
merge_df
```

```python
merge_df.set_index("구 별", inplace=True)
```

```python
merge_df["1인당 cctv 비율"] = ( merge_df["cctv 총계"] / merge_df["총인구"] ) * 100
merge_df.head(3)
```

```python
merge_df["총인구"].corr(merge_df["cctv 총계"],method='spearman')
```

```python
merge_df["외국인"].corr(merge_df["cctv 총계"],method='spearman')
```

#### 시각화

```python
plt.figure(figsize=(12,8))
merge_df.sort_values("1인당 cctv 비율",inplace=True)
plt.barh(merge_df.index,merge_df["1인당 cctv 비율"])
plt.xlabel("cctv 비율")
plt.ylabel("구 별")
plt.title("구 별 1인당 cctv 비율")
plt.show()
```

```python
merge_df['1인당 cctv 비율'].plot.barh()
```

```python
sns.barplot(data=merge_df,x="1인당 cctv 비율",y=merge_df.index)
```

```python
crime_df = pd.read_excel("./data/5대범죄_2021.xlsx",usecols='B:D',skiprows=3)
crime_df = crime_df.drop(0, axis=0)
crime_df.head(4)
```

```python
crime_df
```

```python
# 컬럼명 변경
crime_df.columns=["구 별","범죄건","검거건"]
```

```python
crime_df.set_index("구 별", inplace=True)
```

```python
crime_df["검거율"] = crime_df["검거건"] / crime_df["범죄건"] *100
```

```python
crime_df
```

```python
merge_crime_df = pd.concat([merge_df,crime_df], axis=1)
```

```python
merge_crime_df
```

```python
plt.scatter(merge_crime_df["cctv 총계"],merge_crime_df["검거건"])
plt.scatter(merge_crime_df["cctv 총계"],merge_crime_df["범죄건"],color="red")
```
