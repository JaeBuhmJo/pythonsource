```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--start-maximized")
    # --headless : 브라우저는 뜨지 않고 결과만 크롤링해오는 옵션
    # chrome_options.add_argument("--headless")
    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()), options=chrome_options
    )
    return driver
```

```python
# 구글 이미지 다운로드
from urllib.request import urlretrieve

driver = set_chrome_driver()
driver.get("https://www.google.com/imghp?hl=ko&tab=ri&ogbl")

# 
search = driver.find_element(By.XPATH, '//*[@id="APjFqb"]')
search.send_keys("파이썬")
search.send_keys(Keys.ENTER)

prev_height = driver.execute_script("return document.body.scrollHeight")

interval = 2

while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(interval)

    curr_height = driver.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height:
        #결과 더보기 클릭
        seeMore = driver.find_element(By.XPATH,'//*[@id="islmp"]/div/div/div/div/div[1]/div[2]/div[2]/input')
        try:
            seeMore.click()
        except:
            break

    prev_height = curr_height

# 이미지 저장

# 전체 이미지 가져오기
images = driver.find_elements(By.CSS_SELECTOR,"div.bRMDJf.islir > img")
count = 1
for img in images:
    try:
        # 작은 이미지 클릭
        img.click()
        # 큰 이미지 경로 찾기
        imgUrl = driver.find_element(By.XPATH,'//*[@id="Sva75c"]/div[2]/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img[1]').get_attribute("src")
        print(imgUrl)
        # urlretrieve() 저장
        urlretrieve(imgUrl,"./download/"+str(count)+".jpg")
        count+=1
    except:
        pass
```

```python
import requests
from fake_useragent import UserAgent

from io import BytesIO
import time
from openpyxl import Workbook
from openpyxl.drawing.image import Image
from selenium.webdriver.common.by import By # 파싱 기준
from selenium.webdriver.common.keys import Keys # 키보드 키 지정

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

wb = Workbook()
ws = wb.active
ws.title = "다나와 노트북 애플 검색"
ws.column_dimensions["A"].width = 60
ws.column_dimensions["B"].width = 20
ws.append(["제품명","가격","이미지"])

driver = set_chrome_driver()
driver.get("https://prod.danawa.com/list/?cate=112758&15main_11_02")
driver.implicitly_wait(2) # time.sleep(2)

try:
    WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//*[@id="dlMaker_simple"]/dd/div[2]/button[1]')))
    driver.find_element(By.XPATH,'//*[@id="dlMaker_simple"]/dd/div[2]/button[1]').click()
    WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//*[@id="searchMaker1452"]')))
    driver.find_element(By.XPATH,'//*[@id="searchMaker1452"]').click()

    curr_page, target_crawl_num = 1, 4
    #엑셀 행 번호
    ins_cnt = 2
    while curr_page <= target_crawl_num:
        time.sleep(3)

        products = driver.find_elements(By.CSS_SELECTOR, "div.main_prodlist > ul.product_list > li.prod_item")
        for idx, product in enumerate(products):
            if(idx+1!=len(products)):
                prodName = product.find_element(By.CSS_SELECTOR,"p.prod_name a")
                prodPrice = product.find_element(By.CSS_SELECTOR,"div.prod_pricelist p.price_sect strong")
                prodImg = ""
                img = product.find_element(By.CSS_SELECTOR,"div.thumb_image img")
                if(img.get_attribute("data-original")):
                    prodImg = "https:"+str(img.get_attribute("data-original"))
                else:
                    prodImg = img.get_attribute("src")
                # 제품명, 가격, 이미지
                ws.append([prodName.text, prodPrice.text])
                res = requests.get(prodImg, headers={"user-agent" : UserAgent().chrome})
                prodImgFile = BytesIO(res.content)
                image = Image(prodImgFile)
                image.width = 30
                image.height = 20
                ws.add_image(image, "C"+str(ins_cnt))
                ins_cnt+=1
                # 엑셀에 이미지 저장
        # 현재 페이지 번호를 변경
        curr_page += 1

        #다음 페이지 번호 클릭
        pageButtons = driver.find_elements(By.CSS_SELECTOR,'div.number_wrap a')
        for pageButton in pageButtons:
            if(pageButton.text == str(curr_page)):
                pageButton.click()
except TimeoutException as e:
    print(e)

wb.save("danawa_apple.xlsx")
wb.close()
```
