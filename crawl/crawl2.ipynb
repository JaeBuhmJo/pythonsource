```python
import requests

r = requests.get("https://jsonplaceholder.typicode.com/todos/1")

print(r.headers)
print(r.encoding)
print(r.json())
print(r.json().keys())
print(r.json().values())
```

```python
r = requests.get("https://jsonplaceholder.typicode.com/users")

for row in r.json():
    for k,v in row.items():
        print(k,":",v)
    print()
```
