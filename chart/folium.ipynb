```python
m = folium.Map(
    location=[37.566697,126.978426],
    tiles="Stamen Terrain",
    zoom_start=15
)

icon_plane = plugins.BeautifyIcon(
    icon="plane", border_color='darkblue',text_color='darkblue',icon_shape='circle'
)

folium.Marker(
    [37.5687865,126.9685111],
    popup="<i>돈의문 박물관</i>",
    icon=folium.Icon(color="red",icon="info-sign"),
).add_to(m)
folium.Marker(
    [37.562402,126.801289],
    popup="<i>김포공항</i>",
    icon=icon_plane,
).add_to(m)

folium.Marker(
    [37.5693255,126.98600066],
    popup="<i>솔데스크</i>",
    icon = folium.Icon(color="blue",icon="star")
).add_to(m)
m
```

```python
m = folium.Map(
    location=[37.566697,126.978426],
    tiles="Stamen Terrain",
    zoom_start=15
)

folium.Marker(
    [37.5687865,126.9685111],
    popup="<i>돈의문 박물관</i>",
).add_to(m)

m.add_child(folium.ClickForMarker(popup="Marker"))
```

```python
m = folium.Map(
    location=[37.566697,126.978426],
    tiles="Stamen Terrain",
    zoom_start=12
)

folium.Circle(
    [37.5687865,126.9685111],
    popup="<i>돈의문 박물관</i>",
    radius=60,
    color="royalblue",
    fill=False
).add_to(m)
m
```

```python
cities = [
    [37.566687,126.978417],
    [37.179774,129.075004],
    [37.455900,128.705522],
    [37.350451,127.384827],
    [35.160072,126.851440],
]

m = folium.Map(
    location=[37.566697,126.978426],
    zoom_start=7
)

for i in range(len(cities)):
    folium.Circle(
        location=cities[i],
        radius=200,
        color="red",
    ).add_to(m)
m
```
