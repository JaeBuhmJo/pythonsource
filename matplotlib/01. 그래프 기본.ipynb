```python
import matplotlib.pyplot as plt
```

### 기본 차트 생성

><b>리스트를 그래프로 나타내기</b>

```python
list1 = [10,20,30,40]

# 차트 함수 : plot()
plt.plot(list1)
```

```python
x = [1,2,3,4]
y = [12,43,25,15]

plt.title("선형 차트")
g = plt.plot(x,y)
```

><b>파이썬 딕셔너리를 그래프로 나타내기</b>

```python
dict1 = {"data_x":[1,2,3,4,5],"data_y":[2,3,5,7,9]}
plt.plot("data_x","data_y",data=dict1)
plt.show()
```
