#### 원 크기 변경 - 학년 값 이용

```python
#
sizes = df1["학년"]*500
plt.scatter(df1["영어"],df1["수학"], s=sizes, cmap="viridis", c=df1["학년"])
plt.xlabel("영어")
plt.ylabel("수학")
plt.colorbar()
plt.show()
```

#### 색상 변경

#### cmap (https://matplotlib.org/stable/tutorials/colors/colormaps.html)

#### alpha : 투명

#### colorbar() 

```python
#
sizes = df1["학년"]*500
plt.scatter(df1["영어"],df1["수학"], s=sizes, cmap="viridis", c=df1["학년"])
plt.xlabel("영어")
plt.ylabel("수학")
plt.colorbar(ticks=[1,2,3],label="학년",shrink=0.8,orientation="horizontal")
plt.show()
```

- ticks : 기본으로 나오는 눈금 대신 다른 것 지정
- label : 컬러바 라벨 지정
- shrink : 컬러바 길이 지정
- horizontal : 컬러바 위치(가로)
