#### [실습1]

```python
values = [30,25,20,13,10,2]
labels = ["Python","Java","JavaScript","C#","C/C++","etc"]
```

```python
# plt.plot() : 선차트, bar() : 세로막대차트, barh() : 가로막대차트, pie() : 파이차트
plt.pie(values, labels=labels, autopct="%.1f%%", counterclock=False, startangle=90)
plt.show()
```

- autopct : 원 안에 표시할 % 포맷
- startangle : 안 주는 경우 오른쪽 15도 정도에서 시작해서 시계 반대 방향으로 원이 돌아가게 됨, 각도를 주게 되면 그 각도를 기준으로 시계 반대 방향으로 원이 그려짐
- counterclock=False : 시계 방향으로 원이 작성되도록 함

#### [실습2] 원 조각 내기

```python
# plt.plot() : 선차트, bar() : 세로막대차트, barh() : 가로막대차트, pie() : 파이차트
explode = [0.2,0.1,0,0,0,0]
plt.pie(values, labels=labels, autopct="%.1f%%", counterclock=False, startangle=90,explode=explode)
plt.show()
```

```python
# plt.plot() : 선차트, bar() : 세로막대차트, barh() : 가로막대차트, pie() : 파이차트
explode = [0.05]*6
plt.pie(values, labels=labels, autopct="%.1f%%", counterclock=False, startangle=90,explode=explode)
plt.show()
```

#### [실습3] 범례

```python
# plt.plot() : 선차트, bar() : 세로막대차트, barh() : 가로막대차트, pie() : 파이차트
explode = [0.05]*6
plt.pie(values, labels=labels, autopct="%.1f%%", counterclock=False, startangle=90,explode=explode)
plt.legend()
plt.show()
```

## 심화

### 1) 색상 변경

### 2) wedgeprops() : 도넛 모양 형태로 구성

```python
# plt.plot() : 선차트, bar() : 세로막대차트, barh() : 가로막대차트, pie() : 파이차트
wedgeprops = {"width":0.8, "edgecolor":"w","linewidth":3}
plt.pie(values, labels=labels, autopct="%.1f%%", counterclock=False, startangle=90,wedgeprops=wedgeprops)
plt.legend()
plt.show()
```

#### explode 대신에 wedgeprops 이용

### 3) autopct 커스텀

```python
def custom_autopct(pct):
    return ("%.1f%%" % pct) if pct >= 10 else ""
```

```python
plt.pie(values, labels=labels, autopct=custom_autopct, counterclock=False, startangle=90,wedgeprops=wedgeprops)
plt.show()
```
